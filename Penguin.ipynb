```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Plot the count of species
sns.countplot(x='species', data=penguins, hue='species', palette='Set2')
plt.title('Distribution of Penguins Species')
plt.show()
```

```python
# Scatter plot between flipper length and body mass by species
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', data=penguins, hue='species', palette='coolwarm')
plt.title('Flipper Length vs Body Mass by Species')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.show()
```

```python
# Boxplot of body mass across different species by island
sns.boxplot(x='species', y='body_mass_g', data=penguins, hue='island', palette='muted')
plt.title('Body Mass Distribution Across Species by Island')
plt.show()
```

```python
# Pairplot with species as hue
sns.pairplot(penguins, hue='species', palette='pastel')
plt.suptitle('Pairplot of Penguins Dataset', y=1.02)
plt.show()
```

```python
#Heatmap of correlation matrix
# # Select only the numerical columns
numerical_data = penguins.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr = numerical_data.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
