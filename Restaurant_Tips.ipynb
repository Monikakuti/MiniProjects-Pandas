```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset('tips') #inbuild dataset 
tips.head()
```

```python
# Show total tip amount by gender
sns.barplot(x='sex', y='tip', data=tips, estimator=sum)
plt.title('Total Tips by Gender')
plt.show()
```

```python
# Scatter plot between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
# Average tip by smoker/non-smoker
sns.barplot(x='smoker', y='tip', data=tips)
plt.title('Average Tip by Smoking Status')
plt.show()
```

```python
# Total sales (total bill) by day
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('Total Sales by Day')
plt.show()
```

```python
# Average tip during Lunch vs Dinner
sns.barplot(x='time', y='tip', data=tips)
plt.title('Average Tip by Time of Day')
plt.show()
```

```python
# Effect of group size on tip
sns.barplot(x='size', y='tip', data=tips)
plt.title('Average Tip by Group Size')
plt.show()
```

```python
# Gender vs Tip Percentage (who tips more generously?)
tips['tip_percent'] = (tips['tip'] / tips['total_bill']) * 100
sns.barplot(x='sex', y='tip_percent', data=tips)
plt.title('Average Tip Percentage by Gender')
plt.show()
```

```python
# Tip by Day and Smoking Status
sns.barplot(x='day', y='tip', hue='smoker', data=tips)
plt.title('Tip by Day and Smoking Status')
plt.show()
```

```python
# Customer count by day (busiest day)
sns.countplot(x='day', data=tips)
plt.title('Number of Customers by Day')
plt.show()
```

```python
# Distribution of Tips
sns.histplot(tips['tip'], bins=20, kde=True)
plt.title('Distribution of Tips')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Frequency')
plt.show()
```
